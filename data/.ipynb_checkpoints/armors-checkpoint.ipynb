```python
import pandas
```

```python
import requests
```

```python
url = 'https://finalfantasy.fandom.com/wiki/Final_Fantasy_VII_armor'
```

```python
r = requests.get(url)
```

```python
dfs = pandas.read_html(r.text, attrs={'class': 'full-width FFVII article-table'})
```

```python
df = dfs[0]
```

```python
df = df.drop(['Materia', 'Acquire', 'Unnamed: 8'], axis='columns')
```

```python
df
```

```python
df.drop_duplicates('Armor', inplace=True)
```

```python
df
```

```python
import re
```

```python
re.sub(r'^([a-zA-Z\s]+)(\(|\[).*$', r'\g<1>', 'Precious Watch(Mythril[note 4])')
```

```python
df['Armor'].apply(lambda armor: re.sub(r'^([a-zA-Z\s]+)(\(|\[).*$', r'\g<1>', armor))
```

```python
df['Armor'] = df['Armor'].apply(lambda armor: re.sub(r'^([a-zA-Z\s]+)(\(|\[).*$', r'\g<1>', armor))
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df.drop('index', axis='columns', inplace=True)
```

```python
df.columns = map(str.lower, df.columns)
```

```python
df
```

```python
df.to_csv("~/college/mobile-dev/semester_project/data/armors.csv")
```
